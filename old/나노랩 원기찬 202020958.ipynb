# Step 4: Query 임베딩 생성

```python
try:
    query = "해물탕 만드는거 알려줘"
    embedded_query = query_embeddings.embed_query(query)
    print(f"[INFO] Query Embedding 생성 성공. 임베딩 차원: {len(embedded_query)}")
except Exception as e:
    print(f"[ERROR] Query Embedding 생성 실패: {e}")
```

# Step 5: Passage 임베딩 생성

```python
try:
    desc = texts["Description"]
    embedded_documents = passage_embeddings.embed_documents(desc)
    print(f"[INFO] Passage Embedding 생성 성공. 임베딩 수: {len(embedded_documents)}")
except Exception as e:
    print(f"[ERROR] Passage Embedding 생성 실패: {e}")
```

# Step 6: 유사도 계산

```python
try:
    similarity = np.array(embedded_query) @ np.array(embedded_documents).T
    sorted_idx = similarity.argsort()[::-1]  # 높은 유사도 순으로 정렬
except Exception as e:
    print(f"[ERROR] 유사도 계산 실패: {e}")
```

# Step 7: 결과 출력

```python
print("\n[Query]")
print(f"====================================\n{query}\n")
print(f"{texts['Title'][sorted_idx[0]]} | Description: {texts['Description'][sorted_idx[0]]}")
print("====================================")
print("[Top 5 Results]")
print("====================================")
for i, idx in enumerate(sorted_idx[:5]):
    print(f"[{i+1}] 유사도: {similarity[idx]:.3f} | Title: {texts['Title'][idx]} | Description: {texts['Description'][idx]}")
```
